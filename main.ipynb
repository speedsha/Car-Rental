```python
from car_rental import CarRental, Customer

def main():
    rental = CarRental(20)
    customer = Customer()

    while True:
        print("""
===== CAR RENTAL SYSTEM =====
1. Display Available Cars
2. Rent Car Hourly
3. Rent Car Daily
4. Rent Car Weekly
5. Return Cars
6. Exit
-----------------------------
""")

        choice = input("Enter your choice: ")

        if choice == '1':
            rental.display_stock()

        elif choice == '2':
            cars = customer.request_cars()
            if cars:
                customer.rental_time = rental.rent_hourly(cars)
                customer.rental_mode = 1

        elif choice == '3':
            cars = customer.request_cars()
            if cars:
                customer.rental_time = rental.rent_daily(cars)
                customer.rental_mode = 2

        elif choice == '4':
            cars = customer.request_cars()
            if cars:
                customer.rental_time = rental.rent_weekly(cars)
                customer.rental_mode = 3

        elif choice == '5':
            req = customer.return_cars()
            if req:
                rental.return_cars(req)
                customer.rental_time = None
                customer.rental_mode = None
                customer.cars = 0

        elif choice == '6':
            print("Thank you for using the Car Rental System!")
            break

        else:
            print("Invalid choice! Enter 1â€“6.")

main()
```
